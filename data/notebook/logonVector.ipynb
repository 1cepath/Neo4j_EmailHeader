```python
import os, sys, math
from datetime import datetime, timedelta
```

```python
filepath = 'logon.CSF0929.csv'
```

```python
logfile = open(filepath, 'r')
csvfile = open(filepath.split(".", -1)[0]+".vector.csv",'w')
        
logline = logfile.readline()
pcdic = {}
pccounter = 0
```

```python
while 1:

    if logline == "":
        break
    records = logline.split(",")
    
    startstamp = records[1]
    actioncounter = 1
    pc = 0
    # make id of pc
    if records[3] in pcdic:
        pc = pcdic[records[3]]
    else:
        pcdic[records[3]] = pccounter
        pc = pccounter
        pccounter += 1

    while 1:

        logline = logfile.readline().replace('\r\n', '')
        records = logline.split(",")
        
        print records

        # end condition for current date
        if (logline == "") or (records[1][:10] != startstamp[:10] and records[4] == "Logon"):
            endtime = datetime.strptime(endstamp, "%m/%d/%Y %H:%M:%S")
            starttime = datetime.strptime(startstamp, "%m/%d/%Y %H:%M:%S")
            logontime = (endtime - starttime).total_seconds()
            csvfile.write(str(starttime.hour)+", "+str(endtime.hour)+", "+str(math.floor(logontime/60))+", "+str(actioncounter)+", "+str(pc)+"\n")
            break

        actioncounter += 1
        endstamp = records[1]
```
